## The First Four Moments: Mean, Variance, Skew, and Kurtosis

When we talk about a probablity dist, we often care about at four things:
- The expected value, or where a distribution is centered (mean)
- The amount of variation from that centered value (variation)
- Whether or not our distribution favors, leans towards, or skews towards certain values. (skew)
- How often to expect unusually high or low values. (kurtosis)

**Task3**:

Plot the following distributions: $Beta(1,1)$, $Beta(0.9,0.9)$, $Beta(2,1)$, $Beta(1,2)$, $Beta(7,11)$, $Beta(96,42)$, $Beta(5,5)$, $Beta(500,500)$, $Beta(0.8,0.9)$.

Then, determine which of those distribtuions has:
- the largest/smallest mean
- the largest/smallest variance
- the largest/smallest skew
- the largest/smallest kurtosis

````python
dists=[
    pz.Beta(alpha=1, beta=1),
    pz.Beta(alpha=0.9, beta=0.9),
    pz.Beta(alpha=2, beta=1),
    pz.Beta(alpha=1, beta=2),
    pz.Beta(alpha=7, beta=11),
    pz.Beta(alpha=96, beta=42),
    pz.Beta(alpha=5, beta=5),
    pz.Beta(alpha=500, beta=500),
    pz.Beta(alpha=0.8, beta=0.9)
]
for i, dist in enumerate(dists):
  dist.plot_pdf()
  print(dist.moments())
````

- the largest/smallest mean: Beta(96,42)/Beta(1,2)
- the largest/smallest variance: Beta(0.8,0.9)/Beta(500,500)
- the largest/smallest skew: Beta(1,2)/Beta(2,1)
- the largest/smallest kurtosis: Beta(500,500)/Beta(0.8,0.9)

A Bernoulli distribution is a distribution that graphs the probabality with respect to number of successes, where only one trial is done.

A Binomial distribution is a distribution that graphs the probabality of each number of possible successes, where any number of trials are done.

A Poisson distribution graphs the probability of x events happening in a time window where the average number of events is lambda.

A Beta distribution graphs the likelihood of α successes and β failures occurring with respect to the true probability of a success.

A Geometric distribution graphs the probability of taking x number of attempts to get a success given a probability of success p.

A Hypergeometric distribution graphs the probability of drawing x successes in n draws without replacement from a population of N that contains k successes.

A Categorical distribution gives a probability of the xth element in an array to each x.
